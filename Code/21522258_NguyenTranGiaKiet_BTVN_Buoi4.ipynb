```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
list_edge = [('A', 'B'), ('A', 'C'), ('A', 'D'), ('B', 'D'), ('B', 'E'), 
             ('C', 'D'), ('C', 'F'), ('A', 'F'), ('B', 'G'), ('F', 'D'), 
             ('F', 'G'), ('G', 'D'), ('E', 'D'), ('E', 'G'),
             ('H', 'F'), ('H', 'G'), ('H', 'I'), ('I', 'J')]

# Create the graph
DG = nx.Graph()
DG.add_edges_from(list_edge)

# Specify positions of nodes to match the image
pos = {
    'A': (-1, 1), 'B': (-1, -1), 'C': (0, 2), 'D': (0, 0), 
    'E': (-1, -2), 'F': (1, 1), 'G': (1, -1), 'H': (2, 0), 
    'I': (3, 0), 'J': (4, 0)
}

# Plot the graph
plt.figure(figsize=(8, 8))
plt.axis('off')
nx.draw_networkx(DG, pos=pos, node_size=600, cmap='coolwarm', font_size=14, font_color='white')
plt.show()
```

```python
print("betweenness_centrality:")
print(nx.betweenness_centrality(DG))

print("\ncloseness_centrality:")
print(nx.closeness_centrality(DG))
```
