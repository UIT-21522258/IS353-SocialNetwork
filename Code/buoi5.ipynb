```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
list_edge = [(1,2), (1,5), (2,3), (2,5), (3,4), (4,5), (4,6)]
DG = nx.Graph()
DG.add_edges_from(list_edge)
plt.figure(figsize=(8,8))
plt.subplot(221)
plt.axis('off')
nx.draw_networkx(DG, pos=nx.spring_layout(DG, seed=0), node_size=600, cmap='coolwarm', font_size=14, font_color='white')
```

```python
from networkx.algorithms.community import girvan_newman

communities_generator = girvan_newman(DG)

# Duyệt qua các cấp độ phân rã
for level, communities in enumerate(communities_generator):
    print(f"Level {level}: {list(communities)}")
    
    # Nếu số lượng nhóm đạt đến giá trị mong muốn (ví dụ: mỗi nhóm là một node), dừng lại
    if all(len(c) == 1 for c in communities):
        print(f"Phân rã đến cuối cùng ở cấp độ {level}")
        break
```

```python
part = nx.community.edge_betweenness_partition(DG, 2)
part
```
