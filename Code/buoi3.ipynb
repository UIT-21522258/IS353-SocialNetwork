```python
import networkx as nx
import matplotlib.pyplot as plt

DG = nx.Graph()
# DG.add_edges_from([('A', 'B', {"weight": 10}), ('A', 'C', {"weight": 20}), ('B', 'D', {"weight": 30}), ('B', 'E', {"weight": 40}), ('C', 'F', {"weight": 50}), ('C', 'G', {"weight": 30})])

DG.add_edges_from([
    ('A', 'B'), ('B', 'C'), ('B', 'D'), ('C', 'E'), ('D', 'E'), ('E', 'F'), 
    ('E', 'G'), ('E', 'H'), ('D', 'G'), ('G', 'H')
])

labels = nx.get_edge_attributes(DG, "weight")
plt.axis('off')
nx.draw_networkx(DG, pos=nx.spring_layout(DG, seed=0), node_size=600, cmap='coolwarm', font_size=14, font_color='white')
nx.draw_networkx_edge_labels(DG, pos=nx.spring_layout(DG, seed=0), edge_labels=labels)
```

```python
degree_dict = dict(DG.degree())

# In bậc của mỗi đỉnh
print("Bậc của các đỉnh:")
for node, degree in degree_dict.items():
    print(f"Đỉnh {node}: Bậc {degree}")

# Tính phân phối bậc
degree_distribution = {}
for degree in degree_dict.values():
    if degree in degree_distribution:
        degree_distribution[degree] += 1
    else:
        degree_distribution[degree] = 1

# In phân phối bậc
print("\nPhân phối bậc của đỉnh:")
for degree, count in degree_distribution.items():
    print(f"Bậc {degree}: {count} đỉnh")

nx.draw_networkx(DG, with_labels=True, node_color = 'violet')
```

```python
node_clustering = nx.clustering(DG)
print("Node clustering Coefficients: ")
for node, clutering_coefficient in node_clustering.items():
    print(f"Node {node}: {clutering_coefficient}")

nx.draw_networkx(DG, with_labels=True, node_color = 'green')
```

```python
centrality = nx.degree_centrality(DG)
print("Degree Centrality:", centrality)
nx.draw(DG, with_labels=True, node_color = 'blue')
```

```python
pos = nx.spring_layout(DG, seed=42)

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(DG, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=14, font_weight='bold', arrowstyle='->', arrowsize=20, edge_color='gray')
plt.title("Directed Graph Representation", size=15)
plt.show()
```
