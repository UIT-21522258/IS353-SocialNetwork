```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
list_edge = [(0,1), (0,2),
             (1,2), (1,5),
             (2,3), (2,4),
             (3,4), (3,5),
             (4,5),(4,6),(4,7),
             (5,6),(5,7),
             (6,7)]

DG = nx.Graph()
DG.add_edges_from(list_edge)

pos = {0: (4, 1), 1: (3, 0), 2: (3, 2), 3: (2, 1), 
       4: (1, 2), 5: (1, 0), 6: (0, 2), 7: (0, 0)}

# Vẽ đồ thị
plt.figure(figsize=(8,8))
plt.axis('off')
nx.draw_networkx(DG, pos=pos, node_size=600, cmap='coolwarm', font_size=14, font_color='white')

plt.show()

edge_betweenness = nx.edge_betweenness_centrality(DG, normalized=False)

# In ra chỉ số giữa cho từng cạnh
print("Edge between centrality:")
for edge, betweenness in edge_betweenness.items():
    print(f"{edge}: {betweenness:.4f}")
```

```python
list_edge = [(0,1), (0,2),
             (1,2)]

# Tạo đồ thị
DG_1 = nx.Graph()
DG_1.add_edges_from(list_edge)

pos = nx.spring_layout(DG_1, seed=0)

plt.figure(figsize=(8,8))
plt.axis('off')

nx.draw_networkx_nodes(DG_1, pos, node_size=600, cmap='coolwarm', label=True)

nx.draw_networkx_labels(DG_1, pos, font_size=14, font_color='black')

plt.show()
```
