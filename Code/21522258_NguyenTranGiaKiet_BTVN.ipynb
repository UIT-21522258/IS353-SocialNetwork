```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
DG = nx.Graph()

DG.add_edges_from([(0, 1), (0, 4), (0, 3), (1, 2), (1, 4), (1, 5),
         (2, 5), (2, 6), (3, 4), (3, 7), (4, 5), (4, 7), (4, 8),
         (5, 6), (5, 8), (5, 9), (6, 9), (7, 8), (8, 9)])

labels = nx.get_edge_attributes(DG, "weight")
# Define the custom positions for the nodes
positions = {0: (-1, 1), 1: (0, 1), 2: (1, 1),
             3: (-1.5, 0), 4: (-0.5, 0), 5: (0.5, 0), 6: (1.5, 0),
             7: (-1, -1), 8: (0, -1), 9: (1, -1)}
# plt.axis('off')
plt.figure(figsize=(8, 6))
plt.title("Graph Representation", size=15)
nx.draw_networkx(DG, pos=positions, node_size=600, cmap='coolwarm', font_size=14, font_color='white')
nx.draw_networkx_edge_labels(DG, pos=nx.spring_layout(DG, seed=0), edge_labels=labels)
```

```python
degree_dict = dict(DG.degree())

# In bc ca mi nh
print("Bc ca cc nh:")
for node, degree in degree_dict.items():
    print(f"nh {node}: Bc {degree}")
```

```python
degrees = [deg for node, deg in DG.degree()]

# Plot the degree distribution as a histogram
plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=range(min(degrees), max(degrees) + 2), align='left', rwidth=0.8, color='b')
plt.title("Degree Distribution", size=15)
plt.xlabel("Degree", size=12)
plt.ylabel("Frequency", size=12)
plt.grid(True)

plt.show()
```

```python
node_clustering = nx.clustering(DG)
print("Node clustering Coefficients: ")
for node, clutering_coefficient in node_clustering.items():
    print(f"Node {node}: {clutering_coefficient}")
```
