```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
list_edge = [(1,2), (1,3), (2,4), (3,2), (3,4), (4,5), (4,6)]
```

```python
DG = nx.DiGraph()
DG.add_edges_from(list_edge)
plt.figure(figsize=(8,8))
plt.subplot(221)
plt.axis('off')
nx.draw_networkx(DG, pos=nx.spring_layout(DG, seed=0), node_size=600, cmap='coolwarm', font_size=14, font_color='white')
```

```python
print (f"deg(A) = {DG.degree[2]}")
print (f"deg(A) = {DG.in_degree[2]}")
print (nx.degree_centrality(DG))
print (nx.betweenness_centrality(DG))
print (nx.closeness_centrality(DG))
```
