```python
import matplotlib.pyplot as plt
import networkx as nx

G_0 = nx.Graph()

list_edge = [(0,1), (0,2),
             (1,2), (1,5),
             (2,3), (2,4),
             (3,4), (3,5),
             (4,5),(4,6),(4,7),
             (5,6),(5,7),
             (6,7)]

G_0.add_edges_from(list_edge)

plt.figure(figsize=(21, 16))

plt.subplot(221)

plt.axis('off')

nx.draw_networkx(G_0,pos=nx.spring_layout(G_0,seed=0), node_size=600,cmap='coolwarm',font_size=14,font_color='white')

print(f"Edge between centrality: = {nx.edge_betweenness_centrality(G_0, normalized=False)}")
```
