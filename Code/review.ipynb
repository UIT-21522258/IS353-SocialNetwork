```python
import networkx as nx
import matplotlib.pyplot as plt

# To  th c trng s vi cc cnh t danh sch cnh
edges = [
    ('A', '1', 5), ('A', '2', 6), ('A', '3', 8), ('A', '4', 10),
    ('B', '1', 7), ('B', '2', 5), ('B', '3', 3), ('B', '4', 6),
    ('C', '1', 8), ('C', '2', 9), ('C', '3', 7), ('C', '4', 4)
]

G = nx.Graph()

# Thm cc cnh vo  th
for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

# Ly v tr ca cc nt  v  th r rng
pos = nx.spring_layout(G, seed=42)

# V cc cnh vi trng s l nhn
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, font_size=10, font_weight='bold', edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Hin th  th
plt.title(" th biu din chi ph vn chuyn gia cc nh my v ca hng")
plt.show()
```
